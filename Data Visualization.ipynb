### Plotting from an IPython notebook

`magic command`

-  `%matplotlib notebook` will lead to **interactive** plots embedded within the notebook
-  `%matplotlib inline` will lead to **static** images of your plot embedded in the notebook - `preferred`

`Term "inline", which refers to a setting that allows you to display Matplotlib plots directly in a Jupyter Notebook.`
`However, when you use the "inline" setting, the plots are embedded directly in the Jupyter Notebook`

```python
%matplotlib inline
```

%matplotlib is a magic function in the Python programming language that is used to enable plotting within Jupyter notebooks. This function is used to display and manipulate graphs, charts, and other visualizations directly within the notebook.

**Why is matplotlib inline used?**

Matplotlib inline is used to display plots directly. When using matplotlib inline, you don't need to save plots as image files or open them in a separate window.

There are several benefits to using "Matplotlib inline" in your Python projects:
 
1.`Efficiency:`
By displaying plots directly in the Jupyter Notebook or IPython shell output, you can save time and make your data analysis workflow more efficient.
 
2.`Interactivity:` 
"Matplotlib inline" allows you to interact with your plots directly in the notebook or shell output, such as zooming, panning, and saving.
 
3.`Ease of use:` 
Using "Matplotlib inline" is straightforward and requires only a single line of code to enable.
 
4.`Reproducibility:` 
When you use "Matplotlib inline", your plots are saved as part of the notebook or shell output, which makes it easier to reproduce your analysis and share your results with others.

Magic functions are a special type of function in IPython that allow for advanced functionality and convenient shortcuts within the interactive Python environment. These functions are preceded by a "%" sign, which tells IPython to treat them as special commands rather than regular Python functions. 

There are two types of magic functions in IPython:-

1.Line magics 
Line magics are magic functions that operate on a single line of code. For example, %timeit is a line magic that can be used to time the execution of a single line of code.
 
2.Cell magics
Cell magics, on the other hand, are magic functions that operate on entire cells of code. For example, %%timeit is cell magic that can be used to time the execution of an entire cell of code.

```python
# Here is an example of using "Matplotlib inline" to display a simple line plot:-
import matplotlib.pyplot as plt
%matplotlib inline
 
x = [1, 2, 3, 4, 5]
y = [3, 6, 9, 12, 15]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")

plt.show()
```

### Extract Year, Year-Month (2015-11) & also Year-Week (2015-50) from Order Date

`Year-Week: 2015-50`

`Year-Month: 2015-11`

`Year: 2015`

`.dt.strftime('%Y-%m')`

```python
df_store['Year-Month'] = df_store['Order Date'].dt.strftime('%Y-%m')  # strftime:- string format time
df_store['Year'] = df_store['Order Date'].dt.strftime('%Y')
df_store['Year-Week'] = df_store['Order Date'].dt.strftime('%Y-%U')
```

```python
df_store.head(2).T
```

```python
df_store.columns
```

```python
df_store.info() 
# here year,year-month,year-week in the object form
```

### Year - Month   all in all

```python
df_store_agg_ym = df_store.groupby('Year-Month')['Sales'].sum().reset_index()
```

```python
### Canvas
plt.subplots(figsize = (14,8))

#plot the line chart
plt.plot(df_store_agg_ym['Year-Month'], df_store_agg_ym['Sales'])
plt.xticks(rotation=45)
plt.show()
```

### Multiple requirements

```python
filt_year=df_store['Year']>='2018'
```

```python
df_store_agg_yw = df_store[filt_year].groupby('Year-Week')[['Sales', 'Profit']].sum().reset_index()
```

```python
df_store_agg_yw
```

```python
## Canvas
plt.subplots(figsize = (14,6)) #plot multiple charts within one canvas
#plot the line chart for sales                                                       # for colors use hex color codes website
plt.plot(df_store_agg_yw['Year-Week'], df_store_agg_yw['Sales'], color='#8E44AD', linewidth = 2, linestyle='-', marker='o',label='WoW sells in $')

#plot the line chart for profit                                                 
plt.plot(df_store_agg_yw['Year-Week'], df_store_agg_yw['Profit'], color='green', linewidth = 2, linestyle='--', marker='o',label='WoW profit in $')

plt.legend()
plt.title('$ Sales Trend Week Over Week')
plt.xlabel('Year-Week - 2018 only')
plt.ylabel('Sales in $')
plt.xticks(rotation=45)                         #rotate x-axis labels
plt.show()
```

```python
## Canvas
plt.subplots(figsize = (14,6)) #plot multiple charts within one canvas
#plot the line chart for sales                                                       # for colors use hex color codes website
plt.plot(df_store_agg_yw['Year-Week'], df_store_agg_yw['Sales'], color='#8E44AD', linewidth = 2, linestyle='-', marker='o',label='WoW sells in $')

#plot the line chart for profit                                                 
plt.fill_between(df_store_agg_yw['Year-Week'], df_store_agg_yw['Profit'], color='#FF80FF', linewidth = 2, linestyle='--',label='WoW profit in $')

plt.legend()
plt.title('$ Sales Trend Week Over Week')
plt.xlabel('Year-Week - 2018 only')
plt.ylabel('Sales in $')
plt.xticks(rotation=45)                         #rotate x-axis labels
plt.show()
```
